```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA 
import seaborn as sns
```

```python
data=pd.read_csv('wine.csv')
data
```

# EDA

```python
data.describe()
```

```python
data.info()
```

```python
data1=data.iloc[:,1:]
data1
```

```python
# Converting data to numpy array
dataa=data1.values
dataa
```

```python
# Normalizing the numerical data 
data_norm=scale(dataa)
data_norm
```

```python
# Applying PCA Fit Transform to dataset
pca=PCA(n_components=13)
pca_values=pca.fit_transform(data_norm)
pca_values
```

```python
# PCA Components matrix or covariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var=pca.explained_variance_ratio_
var 
```

```python
# Cummulative variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
# Final Dataframe
final_df=pd.concat([data['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# Visualization of PCAs
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

# Checking with other Clustering Algorithms
1. Hierarchical Clustering

```python
# Import Libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# As we already have normalized data, create Dendrograms
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(data_norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(data_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
data2=data.copy()
data2['clustersid']=hclusters.labels_
data2
```

# 2. K-Means Clustering

```python
# Import Libraries
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(data_norm)
    wcss.append(kmeans.inertia_)
    
    # Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Build Cluster algorithm using K=3

```python
# Cluster algorithm using K=3
clusters=KMeans(3,random_state=30).fit(data_norm)
clusters
```

```python
clusters.labels_
```

```python
# Assign clusters to the data set
data3=data.copy()
data3['clustersid']=clusters.labels_
data3
```
